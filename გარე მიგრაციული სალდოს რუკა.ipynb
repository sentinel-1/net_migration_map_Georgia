- **Stateless** (рЃЏрЃЮрЃљрЃЦрЃџрЃљрЃЦрЃћрЃЮрЃЉрЃўрЃА рЃљрЃарЃЏрЃЦрЃЮрЃюрЃћ рЃърЃўрЃарЃћрЃЉрЃў) рЃЏрЃљрЃарЃЌрЃЎрЃБрЃЌрЃ«рЃћрЃЊрЃў: рЃњрЃЋрЃўрЃЕрЃЋрЃћрЃюрЃћрЃЉрЃА рЃњрЃљрЃарЃћ рЃЏрЃўрЃњрЃарЃљрЃфрЃўрЃБрЃџ рЃАрЃљрЃџрЃЊрЃЮрЃА рЃўрЃАрЃћрЃЌрЃў рЃърЃўрЃарЃћрЃЉрЃўрЃАрЃЌрЃЋрЃўрЃА рЃЋрЃўрЃАрЃљрЃф рЃљрЃа рЃњрЃљрЃљрЃЕрЃюрЃўрЃљрЃЌ рЃљрЃарЃфрЃћрЃарЃЌрЃў рЃЦрЃЋрЃћрЃДрЃюрЃўрЃА рЃЏрЃЮрЃЦрЃљрЃџрЃљрЃЦрЃћрЃЮрЃЉрЃљ.
- **Not stated** (рЃљрЃБрЃдрЃюрЃўрЃерЃюрЃљрЃЋрЃў) рЃЏрЃљрЃарЃЌрЃЎрЃБрЃЌрЃ«рЃћрЃЊрЃў: рЃњрЃЋрЃўрЃЕрЃЋрЃћрЃюрЃћрЃЉрЃА рЃњрЃљрЃарЃћ рЃЏрЃўрЃњрЃарЃљрЃфрЃўрЃБрЃџ рЃАрЃљрЃџрЃЊрЃЮрЃА рЃўрЃАрЃћрЃЌрЃў рЃърЃўрЃарЃћрЃЉрЃўрЃАрЃЌрЃЋрЃўрЃА рЃарЃЮрЃЏрЃћрЃџрЃЌрЃљ рЃЏрЃЮрЃЦрЃљрЃџрЃљрЃЦрЃћрЃЮрЃЉрЃўрЃА рЃўрЃюрЃцрЃЮрЃарЃЏрЃљрЃфрЃўрЃљрЃф рЃЏрЃЮрЃюрЃљрЃфрЃћрЃЏрЃћрЃЉрЃерЃў рЃљрЃа рЃљрЃарЃўрЃА рЃљрЃдрЃюрЃўрЃерЃюрЃБрЃџрЃў.
- **Other** (рЃАрЃ«рЃЋрЃљ) рЃЏрЃљрЃарЃЌрЃЎрЃБрЃЌрЃ«рЃћрЃЊрЃў: рЃњрЃЋрЃўрЃЕрЃЋрЃћрЃюрЃћрЃЉрЃА рЃњрЃљрЃарЃћ рЃЏрЃўрЃњрЃарЃљрЃфрЃўрЃБрЃџ рЃАрЃљрЃџрЃЊрЃЮрЃА рЃўрЃАрЃћрЃЌрЃў рЃърЃўрЃарЃћрЃЉрЃўрЃАрЃЌрЃЋрЃўрЃА рЃарЃЮрЃЏрЃџрЃћрЃЉрЃўрЃф рЃљрЃарЃўрЃљрЃю рЃарЃЮрЃЏрЃћрЃџрЃўрЃЏрЃћ рЃЎрЃЮрЃюрЃЎрЃарЃћрЃбрЃБрЃџрЃў рЃЦрЃЋрЃћрЃДрЃюрЃўрЃА рЃЏрЃЮрЃЦрЃљрЃџрЃљрЃЦрЃћрЃћрЃЉрЃў, рЃЌрЃБрЃЏрЃфрЃљ рЃЏрЃљрЃЌрЃў рЃЦрЃЋрЃћрЃДрЃљрЃюрЃљ рЃљрЃа рЃљрЃарЃўрЃА рЃАрЃљрЃЦрЃАрЃбрЃљрЃбрЃўрЃА рЃЏрЃўрЃћрЃа рЃњрЃљрЃАрЃљрЃ»рЃљрЃарЃЮрЃЋрЃћрЃЉрЃБрЃџрЃў рЃЦрЃЋрЃћрЃДрЃюрЃћрЃЉрЃўрЃА рЃАрЃўрЃљрЃерЃў рЃЊрЃљ рЃљрЃЏрЃўрЃбрЃЮрЃЏ рЃљрЃарЃўрЃљрЃю рЃљрЃЏ "рЃАрЃ«рЃЋрЃљ" рЃЎрЃљрЃбрЃћрЃњрЃЮрЃарЃўрЃљрЃерЃў рЃЌрЃљрЃЋрЃЏрЃЮрЃДрЃарЃўрЃџрЃў.
- **Total** (рЃ»рЃљрЃЏрЃў) рЃЏрЃљрЃарЃЌрЃЎрЃБрЃЌрЃ«рЃћрЃЊрЃў: рЃњрЃЋрЃўрЃЕрЃЋрЃћрЃюрЃћрЃЉрЃА рЃ»рЃљрЃЏрЃБрЃа рЃњрЃљрЃарЃћ рЃЏрЃўрЃњрЃарЃљрЃфрЃўрЃБрЃџ рЃАрЃљрЃџрЃЊрЃЮрЃА рЃЏрЃЮрЃфрЃћрЃЏрЃБрЃџрЃў рЃърЃћрЃарЃўрЃЮрЃЊрЃўрЃА рЃњрЃљрЃюрЃЏрЃљрЃЋрЃџрЃЮрЃЉрЃљрЃерЃў рЃЊрЃљ рЃљрЃЦ рЃерЃћрЃАрЃБрЃџрЃўрЃљ рЃДрЃЋрЃћрЃџрЃљ рЃЏрЃўрЃњрЃарЃљрЃюрЃбрЃўрЃА рЃ»рЃљрЃЏрЃБрЃарЃў рЃАрЃбрЃљрЃбрЃўрЃАрЃбрЃўрЃЎрЃљ рЃЏрЃљрЃЌрЃў рЃЏрЃЮрЃЦрЃљрЃџрЃљрЃЦрЃћрЃЮрЃЉрЃўрЃА рЃЏрЃўрЃБрЃ«рЃћрЃЊрЃљрЃЋрЃљрЃЊ.

# рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮрЃА рЃњрЃљрЃарЃћ рЃЏрЃўрЃњрЃарЃљрЃфрЃўрЃБрЃџрЃў рЃАрЃљрЃџрЃЊрЃЮ

рЃњрЃљрЃарЃћ рЃЏрЃўрЃњрЃарЃљрЃфрЃўрЃБрЃџрЃў рЃАрЃљрЃџрЃЊрЃЮ рЃЊрЃљрЃЊрЃћрЃЉрЃўрЃЌрЃў рЃарЃўрЃфрЃ«рЃЋрЃўрЃљ. рЃЌрЃљрЃЋрЃљрЃЊ рЃцрЃћрЃарЃўрЃА рЃўрЃюрЃбрЃћрЃюрЃАрЃўрЃЋрЃЮрЃЉрЃљ рЃЎрЃў рЃЮрЃарЃўрЃЋрЃћ рЃцрЃћрЃарЃўрЃА рЃерЃћрЃЏрЃЌрЃ«рЃЋрЃћрЃЋрЃљрЃерЃў рЃЏрЃўрЃњрЃЋрЃљрЃюрЃўрЃерЃюрЃћрЃЉрЃА рЃерЃћрЃАрЃљрЃЉрЃљрЃЏрЃўрЃАрЃў рЃарЃўрЃфрЃ«рЃЋрЃўрЃА рЃАрЃўрЃЊрЃўрЃЊрЃћрЃА (рЃћ.рЃў. рЃарЃљрЃф рЃБрЃцрЃарЃЮ рЃўрЃюрЃбрЃћрЃюрЃАрЃўрЃБрЃарЃўрЃљ рЃцрЃћрЃарЃў, рЃЏрЃўрЃЌ рЃБрЃцрЃарЃЮ рЃЊрЃўрЃЊрЃўрЃљ рЃерЃћрЃАрЃљрЃЉрЃљрЃЏрЃўрЃАрЃў рЃАрЃбрЃљрЃбрЃўрЃАрЃбрЃўрЃЎрЃБрЃарЃў рЃЏрЃљрЃЕрЃЋрЃћрЃюрЃћрЃЉрЃћрЃџрЃў). рЃЦрЃЋрЃћрЃДрЃюрЃћрЃЉрЃўрЃАрЃЌрЃЋрЃўрЃА рЃцрЃћрЃарЃћрЃЉрЃўрЃА рЃЏрЃўрЃюрЃўрЃГрЃћрЃЉрЃљ (рЃгрЃўрЃЌрЃџрЃљрЃЊ рЃўрЃЦрЃюрЃћрЃЉрЃљ рЃЏрЃЮрЃюрЃўрЃерЃюрЃБрЃџрЃў рЃЌрЃБ рЃЏрЃгрЃЋрЃљрЃюрЃћрЃЊ) рЃЊрЃљ рЃљрЃАрЃћрЃЋрЃћ рЃЏрЃљрЃЌрЃў рЃцрЃћрЃарЃћрЃЉрЃўрЃА рЃўрЃюрЃбрЃћрЃюрЃАрЃўрЃБрЃарЃЮрЃЉрЃўрЃА рЃерЃћрЃарЃЕрЃћрЃЋрЃљрЃф рЃљрЃарЃўрЃА рЃљрЃЋрЃбрЃЮрЃЏрЃљрЃбрЃБрЃарЃў рЃърЃарЃЮрЃфрЃћрЃАрЃў рЃЊрЃљ рЃЊрЃљрЃЏрЃЮрЃЎрЃўрЃЊрЃћрЃЉрЃБрЃџрЃўрЃљ рЃЏрЃ«рЃЮрЃџрЃЮрЃЊ рЃЊрЃљ рЃЏрЃ«рЃЮрЃџрЃЮрЃЊ рЃерЃћрЃАрЃљрЃЉрЃљрЃЏрЃўрЃА рЃАрЃбрЃљрЃбрЃўрЃАрЃбрЃўрЃЎрЃБрЃа рЃЏрЃЮрЃюрЃљрЃфрЃћрЃЏрЃћрЃЉрЃќрЃћ. рЃљрЃдрЃюрЃўрЃерЃюрЃБрЃџрЃў рЃљрЃЋрЃбрЃЮрЃЏрЃљрЃбрЃБрЃарЃў рЃърЃарЃЮрЃфрЃћрЃАрЃўрЃА рЃбрЃћрЃЦрЃюрЃўрЃЎрЃБрЃарЃў рЃљрЃдрЃгрЃћрЃарЃљ рЃЊрЃљ рЃерЃћрЃАрЃљрЃЉрЃљрЃЏрЃўрЃАрЃў Python-рЃЎрЃЮрЃЊрЃў рЃАрЃарЃБрЃџрЃўрЃљрЃЊ рЃдрЃўрЃљ рЃАрЃљрЃ«рЃўрЃЌ рЃерЃћрЃњрЃўрЃФрЃџрЃўрЃљрЃЌ рЃўрЃ«рЃўрЃџрЃЮрЃЌ рЃљрЃЏрЃљрЃЋрЃћ рЃЊрЃЮрЃЎрЃБрЃЏрЃћрЃюрЃбрЃўрЃА рЃўрЃюрЃњрЃџрЃўрЃАрЃБрЃарЃћрЃюрЃЮрЃЋрЃљрЃю рЃЋрЃћрЃарЃАрЃўрЃљрЃерЃў.



рЃњрЃљрЃюрЃАрЃљрЃќрЃдрЃЋрЃарЃћрЃЉрЃћрЃЉрЃў:
 - **Immigrant** / **рЃўрЃЏрЃўрЃњрЃарЃљрЃюрЃбрЃў** (рЃарЃЮрЃњрЃЮрЃарЃф рЃњрЃљрЃюрЃАрЃљрЃќрЃдрЃЋрЃарЃБрЃџрЃўрЃљ рЃАрЃљрЃЦрЃАрЃбрЃљрЃбрЃўрЃА рЃЏрЃўрЃћрЃа) РђЊ рЃърЃўрЃарЃў, рЃарЃЮрЃЏрЃћрЃџрЃЏрЃљрЃф рЃњрЃљрЃЊрЃЏрЃЮрЃЎрЃЋрЃћрЃЌрЃљ рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮрЃА рЃАрЃљрЃќрЃдрЃЋрЃљрЃарЃў рЃЊрЃљ рЃЉрЃЮрЃџрЃЮ 12 рЃЌрЃЋрЃћрЃерЃў рЃЊрЃљ рЃАрЃБрЃџ рЃЏрЃфрЃўрЃарЃћ 183
   рЃЊрЃдрЃћ (рЃерЃћрЃАрЃљрЃФрЃџрЃћрЃЉрЃћрЃџрЃўрЃљ рЃћрЃА рЃўрЃДрЃЮрЃА рЃарЃљрЃЏрЃЊрЃћрЃюрЃўрЃЏрЃћ рЃерЃћрЃЏрЃЮрЃАрЃЋрЃџрЃўрЃА рЃЎрЃБрЃЏрЃБрЃџрЃљрЃфрЃўрЃБрЃарЃў рЃ»рЃљрЃЏрЃў) рЃўрЃЏрЃДрЃЮрЃцрЃћрЃЉрЃЮрЃЊрЃљ рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮрЃА
   рЃбрЃћрЃарЃўрЃбрЃЮрЃарЃўрЃљрЃќрЃћ рЃЊрЃљ рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮ рЃљрЃа рЃўрЃДрЃЮ рЃЏрЃўрЃАрЃЌрЃЋрЃўрЃА рЃЏрЃБрЃЊрЃЏрЃўрЃЋрЃў рЃАрЃљрЃфрЃ«рЃЮрЃЋрЃарЃћрЃЉрЃћрЃџрЃў рЃЦрЃЋрЃћрЃДрЃљрЃюрЃљ, рЃћ.рЃў. рЃЏрЃљрЃА рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮрЃА
   рЃАрЃљрЃќрЃдрЃЋрЃарЃћрЃЉрЃА рЃњрЃљрЃарЃћрЃЌ рЃгрЃўрЃюрЃљ 12 рЃЌрЃЋрЃћрЃерЃў рЃњрЃљрЃбрЃљрЃарЃћрЃЉрЃБрЃџрЃў рЃљрЃЦрЃЋрЃА рЃАрЃБрЃџ рЃЏрЃфрЃўрЃарЃћ 183 рЃЊрЃдрЃћ.
 - **Emigrant** / **рЃћрЃЏрЃўрЃњрЃарЃљрЃюрЃбрЃў** (рЃарЃЮрЃњрЃЮрЃарЃф рЃњрЃљрЃюрЃАрЃљрЃќрЃдрЃЋрЃарЃБрЃџрЃўрЃљ рЃАрЃљрЃЦрЃАрЃбрЃљрЃбрЃўрЃА рЃЏрЃўрЃћрЃа) РђЊ рЃърЃўрЃарЃў, рЃарЃЮрЃЏрЃћрЃџрЃЏрЃљрЃф рЃЊрЃљрЃбрЃЮрЃЋрЃљ рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮ рЃЉрЃЮрЃџрЃЮ 12 рЃЌрЃЋрЃћрЃерЃў рЃЊрЃљ рЃАрЃБрЃџ рЃЏрЃфрЃўрЃарЃћ 183 рЃЊрЃдрЃћ (рЃерЃћрЃАрЃљрЃФрЃџрЃћрЃЉрЃћрЃџрЃўрЃљ
   рЃћрЃА рЃўрЃДрЃЮрЃА рЃарЃљрЃЏрЃЊрЃћрЃюрЃўрЃЏрЃћ рЃњрЃљрЃАрЃЋрЃџрЃўрЃА рЃЎрЃБрЃЏрЃБрЃџрЃљрЃфрЃўрЃБрЃарЃў рЃ»рЃљрЃЏрЃў) рЃўрЃЏрЃДрЃЮрЃцрЃћрЃЉрЃЮрЃЊрЃљ рЃАрЃ«рЃЋрЃљ рЃАрЃљрЃ«рЃћрЃџрЃЏрЃгрЃўрЃцрЃЮрЃА рЃбрЃћрЃарЃўрЃбрЃЮрЃарЃўрЃљрЃќрЃћ рЃЊрЃљ
   рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮ рЃўрЃДрЃЮ рЃЏрЃўрЃАрЃЌрЃЋрЃўрЃА рЃЏрЃБрЃЊрЃЏрЃўрЃЋрЃў рЃАрЃљрЃфрЃ«рЃЮрЃЋрЃарЃћрЃЉрЃћрЃџрЃў рЃЦрЃЋрЃћрЃДрЃљрЃюрЃљ, рЃћ.рЃў. рЃЏрЃљрЃА рЃЦрЃЋрЃћрЃДрЃюрЃўрЃЊрЃљрЃю рЃњрЃљрЃАрЃЋрЃџрЃљрЃЏрЃЊрЃћ рЃгрЃўрЃюрЃљ 12 рЃЌрЃЋрЃћрЃерЃў
   рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮрЃерЃў рЃњрЃљрЃбрЃљрЃарЃћрЃЉрЃБрЃџрЃў рЃ░рЃЦрЃЮрЃюрЃЊрЃљ рЃАрЃБрЃџ рЃЏрЃфрЃўрЃарЃћ 183 рЃЊрЃдрЃћ (рЃ«рЃљрЃюрЃњрЃарЃФрЃџрЃўрЃЋрЃЮрЃЉрЃўрЃА рЃЎрЃБрЃЏрЃБрЃџрЃљрЃфрЃўрЃБрЃарЃў рЃ»рЃљрЃЏрЃў).
 - **Net migration** / **рЃњрЃљрЃарЃћ рЃЏрЃўрЃњрЃарЃљрЃфрЃўрЃБрЃџрЃў рЃАрЃљрЃџрЃЊрЃЮ** (рЃарЃЮрЃњрЃЮрЃарЃф рЃњрЃљрЃЏрЃЮрЃўрЃДрЃћрЃюрЃћрЃЉрЃљ рЃљрЃЏ рЃЊрЃЮрЃЎрЃБрЃЏрЃћрЃюрЃбрЃерЃў) РђЊ рЃЏрЃЮрЃфрЃћрЃЏрЃБрЃџрЃў рЃЊрЃарЃЮрЃўрЃА рЃърЃћрЃарЃўрЃЮрЃЊрЃерЃў рЃўрЃЏрЃўрЃњрЃарЃљрЃюрЃбрЃЌрЃљ рЃЊрЃљ рЃћрЃЏрЃўрЃњрЃарЃљрЃюрЃбрЃЌрЃљ рЃарЃўрЃфрЃ«рЃЮрЃЋрЃюрЃЮрЃЉрЃљрЃА рЃерЃЮрЃарЃўрЃА рЃАрЃ«рЃЋрЃљрЃЮрЃЉрЃљ.



рЃњрЃљрЃЏрЃЮрЃДрЃћрЃюрЃћрЃЉрЃБрЃџрЃў рЃЏрЃЮрЃюрЃљрЃфрЃћрЃЏрЃћрЃЉрЃў:
- рЃАрЃбрЃљрЃбрЃўрЃАрЃбрЃўрЃЎрЃБрЃарЃў рЃЏрЃЮрЃюрЃљрЃфрЃћрЃЏрЃћрЃЉрЃў "рЃўрЃЏрЃўрЃњрЃарЃљрЃюрЃбрЃћрЃЉрЃўрЃА рЃЊрЃљ рЃћрЃЏрЃўрЃњрЃарЃљрЃюрЃбрЃћрЃЉрЃўрЃА рЃарЃўрЃфрЃ«рЃЮрЃЋрЃюрЃЮрЃЉрЃљ рЃАрЃЦрЃћрЃАрЃўрЃАрЃљ рЃЊрЃљ рЃЏрЃЮрЃЦрЃљрЃџрЃљрЃЦрЃћрЃЮрЃЉрЃўрЃА рЃЏрЃўрЃ«рЃћрЃЊрЃЋрЃўрЃЌ" (Excel рЃцрЃЮрЃарЃЏрЃљрЃбрЃерЃў) рЃЏрЃЮрЃърЃЮрЃЋрЃћрЃЉрЃБрЃџрЃўрЃљ 2022 рЃгрЃџрЃўрЃА 9 рЃўрЃЋрЃюрЃўрЃАрЃА рЃАрЃљрЃЦрЃљрЃарЃЌрЃЋрЃћрЃџрЃЮрЃА рЃАрЃбрЃљрЃбрЃўрЃАрЃбрЃўрЃЎрЃўрЃА рЃћрЃарЃЮрЃЋрЃюрЃБрЃџрЃў рЃАрЃљрЃЏрЃАрЃљрЃ«рЃБрЃарЃўрЃА (рЃАрЃљрЃЦрЃАрЃбрЃљрЃбрЃўрЃА) рЃЋрЃћрЃЉрЃАрЃљрЃўрЃбрЃўрЃЊрЃљрЃю: [geostat.ge](https://www.geostat.ge/ka/)
